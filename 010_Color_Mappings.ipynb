```
import cv2
import matplotlib.pyplot as plt
```

```
img = cv2.imread("/content/1.jpg")
```

```
plt.imshow(img) # 값이 변경되어서 채널이 그래서 색이 다르게나옴
```

```
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img) # 우리가 원하는 색상대로 채널순서를 변경해서 확인하였다.
```
