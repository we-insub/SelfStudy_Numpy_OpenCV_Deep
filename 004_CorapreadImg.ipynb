```
## MATPLOTIB -- > RGB : RED GREEN BLUE
## OPENCV --- > BLUE GREEN RED 순으로 채널을 읽어온다.
```

```
# 아래 셸에서는 BGR 을 RGB 순으로 변경할것이다.
```

```
fix_fig = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #BGR (Blue Green Red) TO RGB (Red Green Blue) 변경
```

```
plt.imshow(img) # 변경전
```

```
plt.imshow(fix_fig) # 변경후 사진을 우리가 평상시보는 원래대로로 바꾼것을 알수 있다.
```

```
w_ratio = 0.5 
h_ratio = 0.5  # 너비와 높이 비율 설정
```

```
new_img = cv2.resize(fix_fig,(0,0),fix_fig,w_ratio,h_ratio) # fix_fig 이미지를 불러와서 0,0 너비 높이 비율 설정
```

```
plt.imshow(new_img) #원래사진 너비가 2500에서 현재는 1250으로 줄었다.
```

```
new_img.shape
```

```
new_img = cv2.flip(fix_fig,0) # 이미지 뒤집기
plt.imshow(new_img) # 이미지 뒤집힌것 확인
```

```
new_img = cv2.flip(fix_fig,1) # 이미지 좌우반전 뒤집기
plt.imshow(new_img) # 이미지 좌우반전 뒤집힌것 확인
```

```
new_img = cv2.flip(fix_fig,-1) # 이미지 좌우반전 뒤집기
plt.imshow(new_img) # 이미지 좌우반전 뒤집힌것 확인
```
