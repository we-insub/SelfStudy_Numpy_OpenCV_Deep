```
# OVERLAY SMALL UIMAGE ON TOP OF A LARGER IMAGE (NO BLENDING)
# Numpy reassignment
```

```
img1 = cv2.imread("/content/1.jpg")
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread("/content/imiage.png")
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
```

```
img2=cv2.resize(img2,(300,300) #이미지 리사이즈 하지만 갖고있는 사진이 워낙작아서 주석처리함
plt.imshow(img2)
```

```
large_img = img1
small_img = img2
```

```
x_offset = 0
y_offset = 0 # 캔버스에 좌표
```

```
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```
small_img.shape
```

```
large_img.shape
```

```
large_img[y_offset:y_end,x_offset:x_end] = small_img
```

```
plt.imshow(large_img) # ㅋㅋㅋㅋㅋ0,0에 완전작게있네...
```
