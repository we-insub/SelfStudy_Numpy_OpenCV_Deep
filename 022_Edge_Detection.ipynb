```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img = cv2.imread("/content/1.jpg")
img = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
plt.imshow(img)
```

```
edges = cv2.Canny(image=img,threshold1=127, threshold2=127)
```

```
plt.imshow(edges)
```

```
edges = cv2.Canny(image=img,threshold1=127, threshold2=255) # 0 -255 값변경시 어떻게 되는지
plt.imshow(edges)
```

```
med_val=np.median(img)
```

```
med_val
```

```
# LOWER THRESHOLD TO EITHER 0 OR 70% OF THE MEDIAN VALUE WICHEVER IS GREATER
lower = int(max(0,0.7*med_val))

# UPPER THRESHOLD TO EITHER 130% OF THE MEDIAN OF THE MAX 255 , WWHICHEVER IS SAMLLER
upper = int(min(255,1.3*med_val))
```

```
edges = cv2.Canny(image=img,threshold1=lower, threshold2=upper) # 0 -255 값변경시 어떻게 되는지
plt.imshow(edges)
```
